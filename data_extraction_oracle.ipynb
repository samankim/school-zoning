```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fields = ['LEAID', 'ALL_MTH00PCTPROF_1516', 'NCESSCH', 'MAM_MTH00PCTPROF_1516', 'MAS_MTH00PCTPROF_1516', 'MBL_MTH00PCTPROF_1516', 'MHI_MTH00PCTPROF_1516', 'MTR_MTH00PCTPROF_1516', 'ECD_MTH00PCTPROF_1516', 'HOM_MTH00PCTPROF_1516', 'MIG_MTH00PCTPROF_1516']
missing_values = ["PS"]
math_achi = pd.read_csv("math-achievement-sch-sy2015-16.csv", usecols = fields, na_values = missing_values, low_memory = False)
math_achi.head()
```

```python
math_achi.replace({'GE':'', 'LE':'', 'GT':'', 'LT':''}, inplace = True, regex = True)
```

```python
for k in range(2, len(math_achi.columns)):
    math_achi.iloc[:, k] = math_achi.iloc[:, k].str.split('-').str[0]
```

```python
math_achi.head()
```

```python
for l in range(len(math_achi.columns)):
    math_achi.iloc[:, l] = pd.to_numeric(math_achi.iloc[:, l], errors='coerce')
math_achi.NCESSCH = math_achi.NCESSCH.astype(float)
```

```python
math_achi['agg_math_achi'] = math_achi.iloc[:, 2:].mean(axis=1)
o_math_achi = math_achi[['LEAID', 'NCESSCH', 'ALL_MTH00PCTPROF_1516']]
```

```python
o_math_achi = o_math_achi.rename(columns={'ALL_MTH00PCTPROF_1516': 'schoolwide_math_achi'})
o_math_achi.head()
```

```python
math_achi.head()
```

```python
fields = ['LEAID', 'NCESSCH', 'ALL_RLA00PCTPROF_1516', 'MAM_RLA00PCTPROF_1516', 'MAS_RLA00PCTPROF_1516', 'MBL_RLA00PCTPROF_1516', 'MHI_RLA00PCTPROF_1516', 'MTR_RLA00PCTPROF_1516', 'ECD_RLA00PCTPROF_1516', 'HOM_RLA00PCTPROF_1516', 'MIG_RLA00PCTPROF_1516']
missing_values = ["PS"]
rla_achi = pd.read_csv("rla-achievement-sch-sy2015-16.csv", usecols = fields, na_values = missing_values, low_memory = False)
rla_achi.head()
```

```python
rla_achi.replace({'GE':'', 'LE':'', 'GT':'', 'LT':''}, inplace = True, regex = True)
```

```python
for k in range(2, len(rla_achi.columns)):
    rla_achi.iloc[:, k] = rla_achi.iloc[:, k].str.split('-').str[0]
```

```python
for l in range(len(rla_achi.columns)):
    rla_achi.iloc[:, l] = pd.to_numeric(rla_achi.iloc[:, l], errors='coerce')
rla_achi.NCESSCH = rla_achi.NCESSCH.astype(float)
```

```python
rla_achi['agg_rla_achi'] = rla_achi.iloc[:, 2:].mean(axis=1)
o_rla_achi = rla_achi[['NCESSCH', 'ALL_RLA00PCTPROF_1516']]
```

```python
o_rla_achi = o_rla_achi.rename(columns={'ALL_RLA00PCTPROF_1516': 'schoolwide_rla_achi'})
o_rla_achi.head()
```

```python
rla_achi.head()
```

```python
o_all_achi = pd.merge(o_math_achi, o_rla_achi, on='NCESSCH', how='inner')
```

```python
o_all_achi.head()
```

```python
o_all_achi['schoolwide_achi'] = o_all_achi.iloc[:, 2:].mean(axis=1)
```

```python
o_all_achi = o_all_achi.dropna(subset=['schoolwide_achi'])
o_all_achi = o_all_achi[['NCESSCH', 'schoolwide_achi']]
o_all_achi = o_all_achi.rename(columns={'NCESSCH': 'school_id'})
```

```python
o_all_achi.to_csv("sw_achi_1516.csv")
```
